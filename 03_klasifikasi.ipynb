```python
# import library dan buat header pada file csv yang dibaca
import pandas as pd
df = pd.read_csv('data_kersen.csv', header=None)
headers = ['H1', 'S1', 'V1', 'H2', 'S2', 'V2','Kelas']
df.columns = headers
df.head()
```

```python
# mengubah isi kolom kelas menjadi angka
df['Kelas'].replace({'Matang':1, 'Mentah':0}, inplace=True)
```

```python
df.head(40)
```

```python
df.dtypes
```

Pembagian data menjadi data latih dan data uji

```python
x = df[['H1', 'S1', 'V1', 'H2', 'S2', 'V2']]
y = df[['Kelas']]
```

```python
from sklearn.model_selection import train_test_split
x_latih, x_uji, y_latih, y_uji = train_test_split(x, y, test_size=0.30, random_state=0)
```

Klasifikasi KNN

```python
from sklearn.neighbors import KNeighborsClassifier
df = KNeighborsClassifier(n_neighbors = 2)
df.fit(x_latih, y_latih.values.ravel())
```

```python
y_prediksi = df.predict(x_uji)
print(y_prediksi)
```

Evaluasi Algoritma KNN

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_uji, y_prediksi)
```

```python
from sklearn.metrics import precision_score
precision_score(y_uji, y_prediksi)
```

```python
from sklearn.metrics import recall_score
recall_score(y_uji, y_prediksi)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_uji, y_prediksi, labels=[1, 0])
```
